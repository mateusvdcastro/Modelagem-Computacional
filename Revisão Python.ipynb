### **O que representa se um trecho de cdigo pertence ou no a uma funo**

Perceba que o cdigo possui uma identao, aqui no jupyter essa indentao  de 2 espaos, nos outros editores o padro  de 4 espaos. \\
Cada indentao, representa a entrada de um contexto. Quando voc indenta para inserir um trecho de cdigo dentro um `if`,  similar a inserir um trecho de cdigo dentro de uma funo. Ento essa estrutura pode ser comparada ao trecho de cdigo dentro de condicionais e loops.

```python
def some_print(a, b, prprint=False):
  soma = a + b

  if prprint:
    print(f'A soma de {a} e {b}  igual a {soma}')
    return
  
  return soma
```

```python
# outro jeito de escrever a funo
def some_print(a, b, prprint=False):
  soma = a + b

  if not prprint:
    return soma
  
  print(f'A soma de {a} e {b}  igual a {soma}')  

# Perceba! Uma funo sem retorno retorna None, e no necessariamente precisamos
# ter argumentos.
  
```

### **Podemos tirar os erros acima usando a keyword global**

Ao usar global dentro de algum escopo, voc define que o escopo que a vriavel (por exemplo) x ir utilizar para leitura e escrita no  o seu escopo local, mas sim o escopo global.

```python
var = 0

def func():
  global var
  print(f'Aps global: var = {var}')
  var = 1
  print(f'Dentro de func(): var = {var}')

func()
print(f'Fora de func(): var = {var}')
```

# **Abrindo e fechando arquivos**

### **`open()`**
A funo `open()` recebe o caminho relativo do arquivo para abrir como argumento. \\
Ou seja, o caminho a partir de onde o terminal que o python est sendo executado est. \\
Alm do caminho para o arquivo, caso o programador queira fazer algo alm de simplesmente ler ele deve modificar o parmetro opcional "mode" (que  o segundo parmetro da funo). \\
Existem os seguintes valores para mode.
1. **`'r'`** (padro): abre um arquivo para leitura
2. **`'w'`**: Abre um arquivo para escrita, se j existir, apaga ele e comea a escrever do 0. Se no existir, o cria
4. **`'a'`**: Abre um arquivo para escrita no final. Se no existir o cria
5. **`'b'`**: Abre um arquivo binrio ao invs de um arquivo de texto
7. **`'+'`**: Combinado com os modos anteriores, atualiza para o modo de atualizao, isto , leitura e escrita

Existem outros modos, mas no abordaremos eles nesse curso. Para saber mais, abra a documentao ou apenas help(open) no python.

A funo retorna um objeto de arquivo.

### **Leitura de arquivos**

### **`arquivo.read()`**

Le o arquivo inteiro e armazena em uma nica string

```python
with open('./lorem.txt') as arq:
  lorem_string = arq.read()

with open('./numeros.txt') as arq:
  numeros_string = arq.read()

print(lorem_string)
print(numeros_string)
```

### **`arquivo.readline()`**
L linha por linha do arquivo, avanando pra prxima linha

```python
with open('./lorem.txt') as arq:
  linha = arq.readline()

  while linha:
    print(linha.strip())
    linha = arq.readline()
```

### **`arquivo.readlines()`**
L o arquivo inteiro, mas armazena um vetor com todas as linhas

```python
with open('./lorem.txt') as arq:
  linhas = arq.readlines()

print(linhas)
print([linha.strip() for linha in linhas])
```

### **Escrita em arquivos**

### **`arquivo.write(string)`**

Escreve no arquivo a string enviada na funo, na posio atual, que pode ser controlada atravs do seek

### **`arquivo.seek(offset: numero)`**

Muda a posio atual do arquivo atravs de um offset de bytes, que  relativo aos valores da posio. \\
Valores: \\
`0` -> Comeo do arquivo \\
`1` -> Posio atual \\
`2` -> Fim do arquivo \\

### **Exemplo: escrevendo em um arquivo (truncando)**

```python
with open('./lorem.txt') as arq:
  lorem_txt = arq.read()

with open('./novo.txt', 'w+') as arq:
  arq.write(lorem_txt)
  
  arq.seek(0)
  print(arq.read())
```

### **Exemplo: adicionando contedo no final do arquivo**

```python
with open('./numeros.txt') as arq:
  numeros_string = arq.read()

with open('./novo.txt', 'a+') as arq:
  arq.write(numeros_string)

  arq.seek(0)
  print(arq.read())
```

### **Slice condicional**

Alm do Slice tradicional, o numpy oferece o slice condicional. Ou seja,  enviado uma condio para realizar um slice no array, **baseado no valor
e no no ndice** \\
Para isso,  necessario fazer um conjunto de condies com operadores no ambguos (`==`, `!=`, `&` para `and`, `|` para `or`).

```python
print(seq[abs(seq) == 100])
```

### **Exemplo: valores diferentes de 0**

```python
print(seq[seq != 0])
```

### **Exemplo: valores maiores que 50**

```python
print(seq[seq > 50])
```

### **Exemplo: valores no intervalo (-inf, -50] ou [50, inf)**

```python
print(seq[(seq <= -50) | (seq >= 50)])
```

### **Exemplo: valores no intervalo [-50, 50]**

```python
print(seq[(seq >= -50) & (seq <= 50)])
```

### **Reproduzindo o operador de adio do NumPy**


```python
array1 = np.array([1, 2, 3, 4])
array2 = np.array([5, 6, 7, 8])

array3 = array1 + array2
array4 = np.concatenate([array1, array2])

list1 = list(array1)
list2 = list(array2)

from operator import add
list3 = list(map(add, list1, list2))
list4 = list1 + list2

print(f'(np.ndarray) array1 + array2: {array3}',
      f'\n(np.ndarray) np.concatenate(array1, array2): {array4}')
print(f'\n(list) Equivalente da soma do NumPy usando map: {list3}',
      f'\n(list) list1 + list2: {list4}')
```

# **Comparativo de performance**

Comparao inspirada em: https://stackoverflow.com/questions/18713321/element-wise-addition-of-2-lists \\

```python
import timeit
from operator import add
import numpy as np

list1 = [4, 5, 6]*10**5
list2 = [1, 2, 3]*10**5

array1 = np.array(list1)
array2 = np.array(list2)

def run_map_add():
  list(map(add, list1, list2))

def run_arr_add():
  array1 + array2

print('Tempo sem NumPy (1 execuo):', timeit.timeit(run_map_add, number=1))
print('Tempo com NumPy (1 execuo):', timeit.timeit(run_arr_add, number=1))
print('Tempo sem NumPy (200 execues):', timeit.timeit(run_map_add, number=200))
print('Tempo com NumPy (200 execues):', timeit.timeit(run_arr_add, number=200))
print('Tempo com NumPy (10000 execues):', timeit.timeit(run_arr_add, number=10000))
```

### **Hello World!**

A fim de termos o primeiro contato com as estruturas de dados bsicas, vamos criar as sries e o dataframe exibidos na imagem ilustrativa e realizar algumas operaes bsicas.

Obs.: Nesse primeiro exerccio tambm iremos realizar a importao da biblioteca.

```python
import pandas as pd # A partir de agora podemos referenciar a biblioteca do Pandas com o "pd"
```

Criando as duas sries com os nomes "Apples" e "Orange"

```python
apples_serie = pd.Series(data=[3,2,0,1], name='Apples')
oranges_serie = pd.Series(data=[0,3,7,2], name='Oranges')

print(apples_serie)
print(oranges_serie)
```

 importante notar que as sries no so simples lista.

```python
type(apples_serie)
```

Os operadores bsicos (+, -, * e /) podem ser utilizados para operar duas sries, gerando uma nova srie. Nesse caso, cada valor da primeira srie  operado com o valor correspondente (mesmo ndice) na segunda srie.

```python
apples_serie + oranges_serie
# apples_serie - oranges_serie
# apples_serie * oranges_serie
# apples_serie / oranges_serie
```

Para criarmos o Dataframe podemos passar as duas sries em formato de dicionrio, onde a **chave** ser o nome da srie (coluna) e o **valor** ser a srie em si.

```python
fruits_df = pd.DataFrame(data={
                        'Apples': apples_serie,
                        'Oranges': oranges_serie
                        })
fruits_df
```

Tambm podemos adicionar sries (colunas)  um dataframe de maneira bastante simples.

```python
total_fruits = apples_serie + oranges_serie
fruits_df['Total Fruits'] = total_fruits
fruits_df
```

# **Explorando um dataset**

A fim de se familiarizar um pouco mais com a estrutura do DataFrame, vamos carregar e trabalhar com um dataset mais interessante.

Todas as funes disponveis no Pandas pode ser encontrada [aqui](https://pandas.pydata.org/docs/reference/frame.html)

Uma grande facilidade oferecida pelo Pandas  o carregamento de dados de diferentes fontes. Para exemplificar isso, alm de termos algo mais interessante para trabalhar, vamos carregar uma base diretamente de uma [planilha](https://docs.google.com/spreadsheets/d/1Bq5IxppilddbsQ0aVAToRkVdeQ0i9KlEsWPLq-fCaQg/edit?usp=sharing) do Google Drive (Retirada do [Kaggle](https://www.kaggle.com/gregorut/videogamesales)).

```python
# No se preocupe com esse trecho de cdigo
# Ele bsicamente vai cuidar da autorizao de acesso e 
# de ler os dados da planilha para a varvel "rows"
# ===========================================================
from google.colab import auth
import gspread
from oauth2client.client import GoogleCredentials
auth.authenticate_user()

GAMES_SALES_DATASET_URL = 'https://docs.google.com/spreadsheets/d/1Bq5IxppilddbsQ0aVAToRkVdeQ0i9KlEsWPLq-fCaQg/edit?usp=sharing'

gc = gspread.authorize(GoogleCredentials.get_application_default())
worksheet = gc.open_by_url(GAMES_SALES_DATASET_URL).sheet1
rows = worksheet.get_all_values()
# ===========================================================

# Vamos dar uma olhada nos dados que esto em "rows"
# Note que  uma simples lista de listas, onde o primeiro item  uma lista com
# os dados de cabealho e os demais itens so as linhas de dados
print(type(rows))
print(rows[0])
print(rows[1])
print(rows[2])
```

Uma vez que os dados da planilha foi carregada para a varivel "rows", vamos utiliza-la para criar um DataFrame do Pandas

```python
# Podemos utilizar a funo from_records() para definir as linhas e cabecalhos
g_sales_df = pd.DataFrame.from_records(columns=rows[0], data=rows[1:])
```

A partir desse ponto, a varivel g_sales_df possui um DataFrame do Pandas com os dados obtidos da planilha referenciada. Vamos investigar um pouco esses dados utilizando algumas funes disponveis.

```python
# A funo head() pode ser utilizada visualizar apenas as primeiras linhas do
# DataFrame

g_sales_df.head() # Por default, so exibidas 5 linhas, mas podemos definir quantas queremos
```

```python
# A funo describe() apresenta um resumo sobre os dados de cada coluna no DataFrame
# As informaes apresentadas variam com o tipo de dados da coluna.

g_sales_df.describe()
```

```python
# Vamos dar uma olhada nos tipos de dados de cada coluna
g_sales_df.dtypes
```

```python
# Visando evitar esse tipo genrico (object), vamos converter as colunas
# para tipos de dados mais interessantes
g_sales_df['Rank'] = g_sales_df['Rank'].astype('int')
g_sales_df['Name'] = g_sales_df['Name'].astype('category')
g_sales_df['Platform'] = g_sales_df['Platform'].astype('category')
g_sales_df['Year'] = g_sales_df['Year'].astype('category')
g_sales_df['Genre'] = g_sales_df['Genre'].astype('category')
g_sales_df['Publisher'] = g_sales_df['Publisher'].astype('category')
g_sales_df['NA_Sales'] = g_sales_df['NA_Sales'].astype('float')
g_sales_df['EU_Sales'] = g_sales_df['EU_Sales'].astype('float')
g_sales_df['JP_Sales'] = g_sales_df['JP_Sales'].astype('float')
g_sales_df['Other_Sales'] = g_sales_df['Other_Sales'].astype('float')
g_sales_df['Global_Sales'] = g_sales_df['Global_Sales'].astype('float')
g_sales_df.dtypes
```

# **Agrupando dados**

Quando trabalhamos com os dados categricos em uma planilha ou em um banco de dados, a realizao de agrupamentos e sumariazaes  bastante comum e til. O Pandas oferece algumas alternativas para esse tipo de operao.

### **Group-By**

Operao similar  opera de mesmo nome utilizada na linguagem SQL para bancos de dados relacionais.

```python
# Obtendo a soma de vendas para cada plataforma
g_sales_df.groupby(by=['Platform']).sum()
```

```python
# Obtendo a mdia de vendas para cada genero
g_sales_df.groupby(by=['Genre']).mean()
```

```python
# Obtendo a soma de vendas para cada genero em cada ano
g_sales_df.groupby(by=['Genre', 'Year']).sum()
```

```python
# Extra - Podemos usar a funo 'loc' para acessar os valores dos anos de um genero especfico
vendas_por_gen_ano = g_sales_df.groupby(by=['Genre', 'Year']).sum()
vendas_por_gen_ano.loc['Racing'].head(10) # Apenas os 10 primeiros
```

### **Pivot Tables**

Formatao bastante utilizada em planilhas.

```python
# Criando uma Pivot table que ir mostras a soma das vendas na North America para cada Ano, Plataforma e Genero
# Obs.: O parmetro fill_value, indica para a funo substituir por 0 qualquer valor faltante
g_sales_df.pivot_table(values='NA_Sales', index=['Year', 'Platform'], columns=['Genre'], aggfunc='sum', fill_value=0)
```

```python
# fig:  a "Figure", basicamente  imagem/figura como um todo e contm os demais elementos.
# ax:  o "Axes", onde a rea de dados  desenhada.
# Importante: "Axes" != "Axis".
# Um "Axes" pode ter 2 ou 3 eixos ("Axis")
# os "Axis" cuida dos limites e ticks

fig, ax = plt.subplots()
ax.plot(range(10), label='linear')
ax.set_xlabel('x label')
ax.set_xlim(left=0, right=15)
ax.set_ylabel('y label')
ax.set_ylim(bottom=0, top=20)
ax.set_title("Grfico Simples")
ax.legend() # Apenas faz a legenda ser exibida, os dados vem do label da curva
ax.grid() # Draw grid lines
# fig.show() Nesse caso, ao invs de plt.show() podemos chamar o show() da "Figure"
```

# **Grficos de Barras**

[API](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.axes.Axes.bar.html#matplotlib.axes.Axes.bar)

```python
# Vamos plotar as vendas de 3 produtos nos quarters de 2019
quarters = ['Q1/2019', 'Q2/2019', 'Q3/2019', 'Q4/2019'] # Eixo x
vendas_pcs = [10, 13, 11, 16]
vendas_laptops = [24, 23, 27, 32]
vendas_video_games = [8, 6, 7, 10]
largura = 0.5

fig, ax = plt.subplots()
ax.bar(x=quarters, height=vendas_pcs, width=largura, label='PCs')

# Para "empilhar" uma barra, marcamos seu inicio no final da anterior com o parmetro "bottom"
ax.bar(x=quarters, height=vendas_laptops, width=largura, bottom=vendas_pcs, label='Laptops')

# Para "empilhar" a terceirca barra, precisamos definir onde as outras ltimas barras terminam
# A funo zip() retorna tuplas com os valores das listas de um mesmo ndice
#print(list(zip(vendas_pcs, vendas_laptops)))
temp_bottom = [a+b for a,b in zip(vendas_pcs, vendas_laptops)]
ax.bar(x=quarters, height=vendas_video_games, width=largura, bottom=temp_bottom, label='Video Games')

ax.set_ylabel('Total de Vendas')
ax.set_xlabel('Quarter')
ax.set_title('Vendas por Quarter e Produto')
ax.legend()
```

# **Multiplos Grficos**

### **Multiplas curvas em 1 grfico**

```python
# Dados para plots
quarters = ['Q1/2019', 'Q2/2019', 'Q3/2019', 'Q4/2019']
vendas_pcs = [10, 13, 11, 16]
vendas_laptops = [24, 23, 27, 32]
vendas_video_games = [8, 6, 7, 10]
bottom_2 = [a+b for a,b in zip(vendas_pcs, vendas_laptops)]
lucro = [3715.76, 3910.23, 4407.53, 6131.59]
largura = 0.5

# Gerando barras
fig, ax_qtd = plt.subplots()
ax_qtd.bar(x=quarters, height=vendas_pcs, width=largura, label='PCs')
ax_qtd.bar(x=quarters, height=vendas_laptops, width=largura, bottom=vendas_pcs, label='Laptops')
ax_qtd.bar(x=quarters, height=vendas_video_games, width=largura, bottom=bottom_2, label='Video Games')

# Gerando linhas
# Gerando novo eixo y com o mesmo x
ax_lucro = ax_qtd.twinx()
ax_lucro.plot(lucro, '--om', label='Lucro', linewidth=4, markersize=8)

# Adicionando Informaes no grfico
ax_qtd.set_xlabel('Quarter')
ax_qtd.set_ylabel('Venda Total')

ax_lucro.set_ylabel('Lucro Total')
ax_lucro.set_ylim(0, 7000)

ax_qtd.set_title('Vendas por Quarter e Produto')
ax_qtd.legend()
ax_lucro.legend(loc='upper center')
fig.show()
```

### **Multiplos grficos em 1 figura**

```python
# Dados para plots
quarters = ['Q1/2019', 'Q2/2019', 'Q3/2019', 'Q4/2019']
vendas_pcs = [10, 13, 11, 16]
vendas_laptops = [24, 23, 27, 32]
vendas_video_games = [8, 6, 7, 10]
products = ['PCs', 'Laptops', 'Video Games']
vendas = [10, 24, 8]
bottom_2 = [a+b for a,b in zip(vendas_pcs, vendas_laptops)]
lucro = [3715.76, 3910.23, 4407.53, 6131.59]
largura = 0.5

# Gerando barras
fig, axs = plt.subplots(2, 2) # 2 linha e 2 colunas

# Quando temos linhas > 1 e colunas > 1, acessamos como uma matriz [x][y]
axs[0][0].bar(x=quarters, height=vendas_pcs, width=largura, label='PCs')
axs[0][0].bar(x=quarters, height=vendas_laptops, width=largura, bottom=vendas_pcs, label='Laptops')
axs[0][0].bar(x=quarters, height=vendas_video_games, width=largura, bottom=bottom_2, label='Video Games')

axs[0][0].set_xlabel('Quarter')
axs[0][0].set_ylabel('Venda Total')
axs[0][0].set_title('Vendas por Quarter e Produto')
axs[0][0].legend()

# Gerando linha
# Gerando novo eixo y com o mesmo x
axs[0][1].plot(quarters, lucro, '--om', label='Lucro', linewidth=4, markersize=8)
axs[0][1].set_ylabel('Lucro Total')
axs[0][1].set_ylim(0, 7000)
axs[0][1].set_title('Lucro')
axs[0][1].legend()

# Gerando pizza
axs[1][0].pie(vendas, labels=products, autopct='%.2f%%', shadow=True)
axs[1][0].set_title('Proporo de venda de produtos')

# Aumentando o tamanho da figura para no distorcer os grficos
fig.set_figwidth(12)
fig.set_figheight(6)

fig.tight_layout() # Adiciona um padding entre os grficos automaticamente

fig.show()
```

# **Pandas + Matplotlib**

O Pandas utiliza o Matplotlib para facilitar a visualizao de seus DataFrames, a seguir vamos investigar um pouco essa integrao.

[API](https://pandas.pydata.org/docs/reference/frame.html#plotting)

```python
# Carregando os dados
import pandas as pd
from google.colab import auth
import gspread
from oauth2client.client import GoogleCredentials
auth.authenticate_user()

GAMES_SALES_DATASET_URL = 'https://docs.google.com/spreadsheets/d/1Bq5IxppilddbsQ0aVAToRkVdeQ0i9KlEsWPLq-fCaQg/edit?usp=sharing'

gc = gspread.authorize(GoogleCredentials.get_application_default())
worksheet = gc.open_by_url(GAMES_SALES_DATASET_URL).sheet1
rows = worksheet.get_all_values()

g_sales_df = pd.DataFrame.from_records(columns=rows[0], data=rows[1:])
g_sales_df['Name'] = g_sales_df['Name'].astype('category')
g_sales_df['Platform'] = g_sales_df['Platform'].astype('category')
g_sales_df['Year'] = g_sales_df['Year'].astype('category')
g_sales_df['Genre'] = g_sales_df['Genre'].astype('category')
g_sales_df['Publisher'] = g_sales_df['Publisher'].astype('category')
g_sales_df['NA_Sales'] = g_sales_df['NA_Sales'].astype('float')
g_sales_df['EU_Sales'] = g_sales_df['EU_Sales'].astype('float')
g_sales_df['JP_Sales'] = g_sales_df['JP_Sales'].astype('float')
g_sales_df['Other_Sales'] = g_sales_df['Other_Sales'].astype('float')
g_sales_df['Global_Sales'] = g_sales_df['Global_Sales'].astype('float')
g_sales_df.head()
```

A funo plot() do DataFrame faz um plot bsico de todas as sries numricas.

```python
g_sales_df.plot()
```

Vamos trabalhar um pouco os dados para melhorar a visualizao.

```python
# Deixando apenas linhas com o valor dos anos
g_sales_df = g_sales_df[g_sales_df['Year'] != 'N/A']
g_sales_df['Year'] = g_sales_df['Year'].astype('int')

# Ordernando pelo ano
g_sales_df = g_sales_df.sort_values(by='Year')

# Tornando a coluna 'Year' como indice
indexed_df = g_sales_df.set_index('Year')

# Removendo coluna Rank
indexed_df = indexed_df.drop(columns='Rank')

indexed_df.head()
```

```python
# Soma de vendas por ano
indexed_df.groupby(by=['Year']).sum().plot()
```

```python
# Soma de vendas por genero
indexed_df.groupby(by=['Genre']).sum().plot.area()
```

```python
# Mdia de vendas por plataforma
indexed_df.groupby(by=['Platform']).mean().plot.bar(figsize=(20, 5))
```

```python
# Relao de vendas entre Plataforma e Genero
indexed_df.plot.scatter(x='Platform', y='Genre', figsize=(20, 5), s='Global_Sales')
```

```python
# Relao de vendas entre North America e Europa
indexed_df.plot.scatter(x='NA_Sales', y='EU_Sales', figsize=(6, 6), s='Global_Sales', xlim=(0, 40), ylim=(0,40))
```
